CAPM CAPM Analysis and Beta Calculation using regression - 
CAPM(Capital Asset Pricing Model) attempts to price securities by examining the relationship that exists between expected returns and risk. 
Read more about CAPM. (Investopedia CAPM reference) 
The Beta of an asset is a measure of the sensitivity of its returns relative to a market benchmark (usually a market index). How sensitive/insensitive is the returns of an asset to the overall market returns (usually a market index like S&P 500 index). What happens when the market jumps, does the returns of the asset jump accordingly or jump somehow? 
Read more about Beta (Investopedia Beta reference)

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
#Importing the gold price csv file
gold = pd.read_csv('GOLD.csv')
gold.fillna(0,inplace=True)
#gold['Pred'].plot(label='Predicted')
#gold['new'].plot(label='New')
#plt.show()
#Fitting a multiple linear regression model 
x = gold[['Open','High','Low','Price']]
y = gold['Pred']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
poly = PolynomialFeatures(degree = 3) 
X_poly = poly.fit_transform(xtrain) 
poly.fit(X_poly, ytrain) 
lin2 = LinearRegression() 
lin2.fit(X_poly, ytrain) 
result = lin2.predict(poly.fit_transform(xtest))
result_2 = lin2.predict(poly.fit_transform(x))
print('The coefficients are:',lin2.coef_)
#Plotting the values to check for descripencies
plt.hist(gold['Pred'],label='Pred')
plt.hist(result,label='new pred')
plt.legend(loc=0)
plt.show()
sb.distplot(gold['Pred'],label='Pred')
sb.distplot(result,label='Pred2')
plt.legend(loc=0)
plt.show()
result_2 = result_2.tolist()
#Filling the pred column using the predicted values
i=0
while(i<(len(result_2))):
    if gold['Pred'][i] == 0.0:
        gold['Pred'][i] = result_2[i]
    i+=1

gold
```

```python
x2 = gold[['Open','High','Low','Price']]
y2 = gold['new']
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(x2, y2, test_size=0.2)
rgrsn2 = LinearRegression()
rgrsn2.fit(xtrain2, ytrain2)
print('The Coefficients are:',rgrsn2.coef_)
result2 = rgrsn2.predict(xtest2)
result2_2 = rgrsn2.predict(x)
plt.hist(gold['new'],label='new')
plt.hist(result2,label='new2')
plt.legend(loc=0)
plt.show()

sb.distplot(gold['new'],label='new')
sb.distplot(result2,label='new2')
plt.legend(loc=0)
plt.show()
result2_2 = result2_2.tolist()
gold['new2'] = result2_2
```

After plotting the regression model we can conclude that the column 'new' is a linear function and column 'Pred' is Polynomial. At first the plot 'Pred' over a linear function can be constructed only to find that it had too much discrepency. Hence we shifted to Polynomial.
```python
#The query is to calculate daily returns for 3 months only
#Had to fetch data for 3months and then calculate returns
igl = pd.read_csv('IGL.csv')
nifty = pd.read_csv('Nifty50.csv')
igl['Date'] = pd.to_datetime(igl['Date'])
nifty['Date'] = pd.to_datetime(nifty['Date'])
slice_igl = igl.loc[igl['Date'] <= '2017-08-15']
slice_nifty = nifty.loc[nifty['Date'] <= '2017-08-15']
igl['Daily Return'] = slice_igl['Close Price'].pct_change(1)
nifty['Daily Return'] = slice_nifty['Close'].pct_change(1)
igl['Monthly Return'] = igl['Close Price'].pct_change(20)
nifty['Monthly Return'] = nifty['Close'].pct_change(20)


from scipy import stats
import statsmodels.api as sm

x_axis = igl['Daily Return'].iloc[1:]
y_axis = nifty['Daily Return'].iloc[1:]
x_axis.fillna(0,inplace=True)
y_axis.fillna(0,inplace=True)
x_axis.drop(x_axis.index[-3:],inplace=True)
#Had to remove last 3 rows as the x and y indices wern't matching
X1 = sm.add_constant(x_axis)
X1.fillna(0,inplace=True)
model = sm.OLS(y_axis, X1)
resultss = model.fit()
print(resultss.summary())
print()
x_axis2 = igl['Monthly Return'].iloc[1:]
y_axis2 = nifty['Monthly Return'].iloc[1:]
x_axis2.fillna(0,inplace=True)
y_axis2.fillna(0,inplace=True)
x_axis2.drop(x_axis2.index[-3:],inplace=True)
X2 = sm.add_constant(x_axis2)
model2 = sm.OLS(y_axis,X2)
resultss2 = model2.fit()
print(resultss2.summary())
```

The beta Value for daily return is positive and is 0>0.0584<1 which means it is less volatile compared to the market and returns are only 5% of what the market has returned in that period. The beta value for monthly return is negative and is -3.063 which means it is negatively correlated to the returns of the market.
